```python
# LABEL ENCODING
```

```python
# converting categorical columns to numericals columns by using encoding method

# importing label encoder from sklearn
from sklearn.preprocessing import LabelEncoder
```

```python
# creating object of LabelEncoder
le = LabelEncoder()
```

```python
# transforming the columns
df1["CarName"] = le.fit_transform(df1["CarName"])
df1["fueltype"] = le.fit_transform(df1["fueltype"])
df1["aspiration"] = le.fit_transform(df1["aspiration"])
df1["doornumber"] = le.fit_transform(df1["doornumber"])
df1["cylindernumber"] = le.fit_transform(df1["cylindernumber"])
df1["carbody"] = le.fit_transform(df1["carbody"])
df1["drivewheel"] = le.fit_transform(df1["drivewheel"])
df1["enginelocation"] = le.fit_transform(df1["enginelocation"])
df1["enginetype"] = le.fit_transform(df1["enginetype"])
df1["fuelsystem"] = le.fit_transform(df1["fuelsystem"])
df1
```

```python
x = df1.drop(["price"], axis = 1)   #Independent Variable
y = df1["price"]                    #Dependent Variable
```

```python
# Splitting data into training and testing
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size = 0.2, random_state = 42)
x_train
```

```python
# Model Creatiom
from sklearn.linear_model import LinearRegression
```

```python
# Creating Model object
lreg = LinearRegression()
```

```python
# Training the model
lreg.fit(x_train, y_train)
```

```python
# Predicting using the trained model
lreg.predict(x_test)
```

```python
y_test.head()
```

```python
lreg.score(x_test, y_test)
```

```python
# Checking for accuracy

from sklearn.metrics import r2_score
y_pred = lreg.predict(x)
y_act = y
r2_score(y_act, y_pred)
```

```python
# Coefficients
lreg.coef_
```

```python
# Intercept
lreg.intercept_
```

```python
# Mean Square Error
from sklearn.metrics import mean_squared_error
mean_squared_error(y_act, y_pred)
```
